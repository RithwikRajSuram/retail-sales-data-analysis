**Histogram of Sales Amounts**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
data = pd.read_csv('/content/retail_sales_dataset.csv')
# Create histogram
plt.hist(data['Total Amount'], bins=10, edgecolor='black')
plt.title("Histogram of total Amounts")
plt.xlabel("total Amount")
plt.ylabel("Frequency")
plt.show()

```

This histogram shows how total transaction amounts are distributed. Most transactions are concentrated in lower sales ranges, while a smaller number of high-value transactions appear as outliers.

**Scatter Plot of Date vs. Sales Amount**

```python
# Convert date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Create scatter plot
plt.scatter(data['Date'], data['Total Amount'], color='blue')
plt.title("Scatter Plot of Date vs. Total Amount")
plt.xlabel("Date")
plt.ylabel("Total Amount")
plt.xticks(rotation=45)
plt.show()
```

This scatter plot shows transaction values across the year and helps identify whether high-value sales occurred during specific time periods.

**Box Plot of Sales Amounts by Store Location**

```python
import seaborn as sns

# Create box plot
sns.boxplot(x='Product Category', y='Total Amount', data=data)
plt.title("Box Plot of Total Amounts by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Amount")
plt.show()
```

This box plot compares transaction amounts across Beauty, Clothing, and Electronics categories to identify spread, median differences, and outliers.
